```python
import numpy as np
import matplotlib.pyplot as nt

jd = 100
x=np.array([i*0.1 + np.random.randn() for i in range(jd)])
y=np.array([i*0.1 for i in range(jd)])

#Buat Fungsi Linier
def fungsi_firman(m,x) :
    y=m*x
    return y

#Prediksi Fungsi Garis
x_prediksi= np.array([0,50])
m_prediksi=1
y_prediksi= fungsi_firman(x_prediksi,m_prediksi)

nt.scatter(x,y)
nt.plot(x_prediksi,y_prediksi,"red")
nt.axis([10,0,10,0])
nt.show
```
