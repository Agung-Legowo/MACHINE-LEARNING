## Matrix Weight

```python
w11 = np.random.uniform(-0.5,0.5)
w12 = np.random.uniform(-0.5,0.5)
w21 = np.random.uniform(-0.5,0.5)
w22 = np.random.uniform(-0.5,0.5)
w31 = np.random.uniform(-0.5,0.5)
w32 = np.random.uniform(-0.5,0.5)
W   = np.array([[w11,w12],[w21,w22],[w31,w32]])
print(W)
```

## Rumus Sigmoid

```python
import numpy as np
def sigmoid (x):
    return 1/(1+np.exp(-x))
```

```python
learning_rate = 0.1
for index, baris in gbd.iterrows():
    w_lama = W
    input = np.array([[baris.x1],[baris.x2]])
    output_hidden = np.dot in (w_lama*input)
    output_learn = sigmoid (output_hidden)
    output_actual = np.array ([[baris.y1],[baris.y2]])
    
    error = output_actual - output_learn
    delta_w = np.dot ((learning_rate * error * output_learn)*(1 - output_learn),(input.T))
    
    w_baru = w_lama - delta_w
    W = w_baru
    tebakan = np.argmax (output_learn)
    jawaban = np.argmax (output_actual)
    label_tebakan = 'Firmansyah Marwan' if tebakan ==1 else  'Abdul Rahman'
    print(f"index= {index} tebakan= {label_tebakan} jawaban= {baris.label}")
```

```python
x1 = 18
x2 = 20
input = np.array([[x1],[x2]])
output_hidden = np.dot(W,input)
output_prediction = sigmoid(output_hidden)
prediction = np.argmax(output_prediction)
label_prediction = 'Firmansyah Marwan' if prediction ==1 else 'Abdul Rahman'
print(f"tinggi = {x1*10} berat={x2*10} adalah {label_prediction}")
```
