```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
scale.fit(de[['Pregnancies', 'Glucose', 'BloodPressure'
              , 'SkinThickness', 'Insulin', 
              'BMI', 'DiabetesPedigreeFunction', 'Age']])
```

```python
X=scale.transform(de[['Pregnancies', 'Glucose', 'BloodPressure'
              , 'SkinThickness', 'Insulin', 
              'BMI', 'DiabetesPedigreeFunction', 'Age']])
```

```python
y=de['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20)
```

```python
import matplotlib.pyplot as plt
```

```python
training_accuracy=[]
test_accuracy=[]
```

```python
model=KNeighborsClassifier(13)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
y_hat=model.predict(x_test)
confusion_matrix(y_test,y_hat)
pd.crosstab(y_test,y_hat,colnames
            =['Hypothesis class'],rownames=['True Class'],margins=True)
```
